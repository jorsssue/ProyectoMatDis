### Número de nodos del grafo

Podemos obtener el numero de los nodos del grafo con la función **G.number_of_nodes()**

```python
G.number_of_nodes()
```

### Grado de cada nodo del grafo

Recordemos que el grado es el numero de aristas que inciden en un nodo.

```python
for i in personajes:
    print(f"Grado de nodo {i}: {G.degree(i)}")
```

### Excentricidades de cada nodo del grafo

Recordemos que la excentricidad es la mayor distancia entre un nodo y otro, y la podemos obtener con ayuda de la función **nx.eccentricity()**

```python
excentricidades = nx.eccentricity(G)
for nodo, i in excentricidades.items():
    print(f"Excentricidad de {nodo}: {i}")
```

### Caminos de dos nodos.

La distancia es la longitud más corta entre dos nodos. La podemos obtener con ayuda de la función **nx.shortest_path(G, source, target)**

```python
Primer_nodo = input("Ingrese el primer nodo: ")
Segundo_nodo = input("Ingrese el segundo nodo: ")

try:
    camino = nx.shortest_path(G, Primer_nodo, Segundo_nodo)
    print(f"La longitud más corta entre {Primer_nodo} y {Segundo_nodo} es: {camino}")
except:
    print(f"No hay un camino entre {Primer_nodo} y {Segundo_nodo}")
```

### Todos los caminos más cortos

Con la función **nx.all_shortest_paths(G, source, target)** nos va arrojar una lista de todos los caminos más cortos entre un nodo y otro.

```python
Primer_nodo = input("Ingrese el primer nodo: ")
Segundo_nodo = input("Ingrese el segundo nodo: ")

try:
    caminos = list(nx.all_shortest_paths(G, Primer_nodo, Segundo_nodo))
    print(f"La longitudes más cortas entre {Primer_nodo} y {Segundo_nodo} son: {caminos}")
except:
    print(f"No hay un camino entre {Primer_nodo} y {Segundo_nodo}")
```

### Radio y Diamétro

Recordemos que el **radio** es la excentricidad mínima de los nodos de un grafo mientras que el **diámetro** es la excentricidad máxima. Con la función **nx.radius** calculamos el radio y con **nx.diameter** el diametro.

```python
Radio = nx.radius(G)
Diametro = nx.diameter(G)

print(f"Radio del grafo: {Radio}")
print(f"Diametro del grafo: {Diametro}")
```

### Centro y Periferia

Debemos recordar que el **centro** es igual al conjunto de nodos cuya excentricidad coincide al **radio** del grafo mientras que la periferia es el conjunto de nodos que es igual al diámetro. Podemos con la función **nx.center(G)** el centro y **nx.periphery** la periferia.

```python
nx.center(G)
```
